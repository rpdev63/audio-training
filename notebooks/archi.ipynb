```python
import keras
import tensorflow as tf

from tensorflow import keras
from keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.layers import GlobalMaxPooling2D, MaxPooling2D, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers, optimizers
import visualkeras
```

```python
model = Sequential()
model.add(Input(shape=(39,174,1)))
model.add(Conv2D(filters=16, kernel_size=(2, 4), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 3)))
model.add(Conv2D(filters=32, kernel_size=(2, 4), activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64, kernel_size=(2, 4), activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=128, kernel_size=(2, 4), activation='relu'))
model.add(GlobalAveragePooling2D())
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))


model2 = Sequential()
model2.add(Input(shape=(39,174,1)))
model2.add(Conv2D(filters=32, kernel_size=(2, 4), activation='relu'))
model2.add(Conv2D(filters=64, kernel_size=(2, 4), activation='relu'))
model2.add(MaxPooling2D(pool_size=2))
model2.add(Conv2D(filters=128, kernel_size=(2, 4), activation='relu'))
model2.add(Conv2D(filters=256, kernel_size=(2, 4), activation='relu'))
model2.add(GlobalAveragePooling2D())
model2.add(Dropout(0.5))
model2.add(Dense(10, activation='softmax'))
```

```python
from PIL import ImageFont
font = ImageFont.truetype("DejaVuSerif.ttf", 12)
```

```python
visualkeras.layered_view(model, draw_volume=True, spacing=50, legend=True, font=font)
```

```python
visualkeras.layered_view(model2, draw_volume=True, spacing=50, legend=True, font=font)
```
